```python
import os
import random
from shutil import copyfile

img_dir = '/content/img_frames'
label_dir = '/content/labels'

output_dir = '/content/data_split'
train_img_output = os.path.join(output_dir, 'train', 'images')
train_label_output = os.path.join(output_dir, 'train', 'labels')
val_img_output = os.path.join(output_dir, 'val', 'images')
val_label_output = os.path.join(output_dir, 'val', 'labels')

os.makedirs(train_img_output, exist_ok=True)
os.makedirs(train_label_output, exist_ok=True)
os.makedirs(val_img_output, exist_ok=True)
os.makedirs(val_label_output, exist_ok=True)

img_files = os.listdir(img_dir)
random.shuffle(img_files)

total_samples = len(img_files)
train_samples = int(total_samples * 0.8)

train_img_files = img_files[:train_samples]
val_img_files = img_files[train_samples:]

for img_file in train_img_files:
    label_file = img_file.replace('.jpg', '.txt')
    img_src = os.path.join(img_dir, img_file)
    label_src = os.path.join(label_dir, label_file)

    if os.path.exists(label_src):
        img_dst = os.path.join(train_img_output, img_file)
        label_dst = os.path.join(train_label_output, label_file)
        copyfile(img_src, img_dst)
        copyfile(label_src, label_dst)
    else:
        print(f"Label file for {img_file} does not exist. Skipping.")

for img_file in val_img_files:
    label_file = img_file.replace('.jpg', '.txt')
    img_src = os.path.join(img_dir, img_file)
    label_src = os.path.join(label_dir, label_file)

    if os.path.exists(label_src):
        img_dst = os.path.join(val_img_output, img_file)
        label_dst = os.path.join(val_label_output, label_file)
        copyfile(img_src, img_dst)
        copyfile(label_src, label_dst)
    else:
        print(f"Label file for {img_file} does not exist. Skipping.")

print("Data splitting and copying complete.")
```
